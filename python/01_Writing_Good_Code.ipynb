# Don´t Repeat Yourself

```python
import pandas as pd

df = pd.read_csv("../data/sdg_literacy_rate.csv")
df.head()
```

```python
df = df.drop(["Country Code","Series Code","Series Name"], axis=1)
df.head()
```

```python
df = df.set_index("Country Name").transpose()
df.head()
```

```python
def process_sdg_data(csv_file, columns_to_drop):
    df = pd.read_csv(csv_file)
    df = df.drop(columns_to_drop, axis=1)
    df = df.set_index("Country Name").transpose()

    return df

df = process_sdg_data("../data/sdg_literacy_rate.csv", ["Series Name", "Series Code", "Country Code"])
df.head()
```

```python
df = process_sdg_data("../data/sdg_electricity_data.csv", ["Series Name", "Series Code", "Country Code"])
df.head()
```

```python
df = process_sdg_data("../data/sdg_urban_population.csv", ["Series Name", "Series Code", "Country Code"])
df.head()
```
